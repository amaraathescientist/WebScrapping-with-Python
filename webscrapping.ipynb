# WebScrapping

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://companiesmarketcap.com/largest-companies-by-revenue/"

page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')

#<table class="default-table table marketcap-table dataTable" style="width:100%">
```

```python
print(soup)
```

```python
table = soup.find('table')
```

```python
titles = table.find_all('th')
```

```python
world_titles = [item.text for item in titles]
```

```python
world_titles
```

```python
df = pd.DataFrame(columns = world_titles)
df2 = pd.DataFrame(columns = world_titles)
df3 = pd.DataFrame(columns = world_titles)
```

```python
df
```

```python
rows = table.find_all('tr')
```

```python
for row in rows[1:18]:
    data = row.find_all('td')
    clean_data = [i.text.strip() for i in data]
    print(clean_data)
    length = len(df)
    df.loc[length] = clean_data
```

```python
df
```

```python
for row in rows[19:77]:
    data = row.find_all('td')
    clean_data2 = [i.text.strip() for i in data]
    print(clean_data2)
    length = len(df2)
    df2.loc[length] = clean_data2
```

```python
df2
```

```python
for row in rows[78:]:
    data = row.find_all('td')
    clean_data3 = [i.text.strip() for i in data]
    print(clean_data3)
    length = len(df3)
    df3.loc[length] = clean_data3
```

```python
df3
```

```python
final_df = pd.concat([df, df2, df3])
```

```python
pd.set_option('display.max.rows', 101)
```

```python
final_df = final_df.set_index('Rank')
```
